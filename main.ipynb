# Detecção de Regiões Verdes Em Imagens de Satélites
## Porposta Utilizando Python e OpenCV 

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# leitura da imagem usando OpenCV
img0 = cv2.imread('satelite/SP santa_rita_do_passaquatro.png'.format(4))
img = img0.copy()

# estudamos histogramas de cores
i = 500
histy = []
histr = []
histg = []
histb = []
img = cv2.medianBlur(img, 31)
for j in range(img.shape[1]):
    r = img[i,j,0]
    g = img[i,j,1]
    b = img[i,j,2]
    y = 0.299 * r + 0.587 * g + 0.114 * b 
    
    histr.append(r)
    histg.append(g)
    histb.append(b)
    
    histy.append(y)

plt.plot(histr, c='r')
plt.plot(histg, c='g')
plt.plot(histb, c='b')
plt.plot(histy, c = "black")
plt.show()
```
